```python
def get_json(path):
    f = open(path, "r")
    data = f.read().replace('\n', ',')
    data = f'[{data}]'
    print(path)
    if data[-2] == ',':
        l = len(data)
        new_data_str = data[:l-2] + "" + data[l-1]
        data = new_data_str
    data_json = json.loads(data)
    f.close()
    return data_json
```

```python
data_json = []

for path in paths:
    data = get_json(path)
    data_json.extend(data)

data = json.load(open('/kaggle/input/data-downloaded/data_downloaded.json'))
data_json.extend(data)
len(data_json)
```

```python
with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(data_json, f, ensure_ascii=False, indent=4)
```
